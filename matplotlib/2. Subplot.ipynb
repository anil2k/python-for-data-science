```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.0, 5.0)
```

```python
x
```

```python
len(x)
```

```python
y1 = np.sin(np.pi*x)
y2 = np.sin(np.pi*x*2)
```

```python
plt.plot(x, y1, 'b--', label='sin(pi*x)')
plt.ylabel('y1 value')
plt.plot(x, y2, 'r--', label='sin(pi*x*2)')
plt.ylabel('y2 value')
plt.xlabel('x value')
plt.title('this is x-y value')
plt.legend()
```

```python
plt.show()
```

```python
plt.subplot(221)
plt.plot(x, y1, 'b--')
plt.subplot(222)
plt.plot(x, y2, 'r--')
plt.subplot(223)
plt.plot(x, y1, 'r*')
plt.subplot(224)
plt.plot(x, y2, 'b*')
```

```python
plt.show()
```

```python
figure, ax = plt.subplots(2,2)
```
